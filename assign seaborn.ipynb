```python
"""Seaborn is a powerful python library for data visualization.
1)scatterplot:scatterplost are used to visualise the relationship between two continuos variables.They can help you to identify patterns, clusters or correlations in your data.(seabornfunction: sns.scatterplot())
2)Histograms: Histograms are used to display the distribution of a single continuous variable. They show the frequency or density of data points within different intervals or bins.Seaborn Function: sns.histplot()
3)Bar plots: Bar plots are used to display categorical data with rectangular bars. They are useful for comparing values of different categories.Seaborn Function: sns.barplot()
4)Box Plots: Box plots are used to visualize the distribution, spread, and skewness of data. They are particularly useful for identifying outliers.Seaborn Function: sns.boxplot()
5) heatmaps: Heatmaps are used to represent data in a 2D matrix where the values are represented as colors. They are great for visualizing correlations or relationships in large datasets.Seaborn Function: sns.heatmap()""" 


```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
```

```python
fmri= sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
```

```python
titanic= sns.load_dataset('titanic')
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
```

```python
diamonds= sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
sns.displot(diamonds['price'])
```

```python
sns.relplot(x='carat',y='price',hue='cut', data=diamonds)
```
